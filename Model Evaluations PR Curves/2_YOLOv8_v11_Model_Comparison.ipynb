```python
import matplotlib.pyplot as plt
from PIL import Image
import os

# --- CONFIGURATION ---
base_path = "."
# List of the 6 baseline model result files
image_files = [
    "yolo8n_BoxPR_curve.png", "yolo8s_BoxPR_curve.png", "yolo8m_BoxPR_curve.png",
    "yolo11n_BoxPR_curve.png", "yolo11s_BoxPR_curve.png", "yolo11m_BoxPR_curve.png"
]

# Create a 2x3 grid (2 rows, 3 columns)
fig, axes = plt.subplots(2, 3, figsize=(24, 16))
axes = axes.flatten() 

for i in range(len(axes)):
    if i < len(image_files):
        file_name = image_files[i]
        full_path = os.path.join(base_path, file_name)
        
        if os.path.exists(full_path):
            img = Image.open(full_path)
            axes[i].imshow(img)
            
            # Formatting title: "yolo8n_BoxPR_curve" -> "YOLO8N BOXPR CURVE"
            clean_title = file_name.replace('.png', '').replace('_', ' ').upper()
            axes[i].set_title(clean_title, fontsize=16, fontweight='bold', pad=10)
            print(f"Included: {file_name}")
        else:
            axes[i].text(0.5, 0.5, f'File not found:\n{file_name}', 
                        ha='center', va='center', color='red', fontweight='bold')
            print(f"Missing: {file_name}")
    
    axes[i].axis('off') 

# Adjust layout to prevent overlapping
plt.tight_layout(pad=5.0)

# Save the final figure for your TFM report
output_name = "YOLO_Baseline_Comparison_Grid.png"
plt.savefig(output_name, dpi=300, bbox_inches='tight')
plt.show()

print(f"\n Grid saved as: {output_name}")
```
